```python
import numpy as np
import math
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
from gurobipy import *
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid", font_scale=1.8, rc={'axes.grid': True, 'figure.dpi': 120})
```

```python
# not tested with bug updates: distance matrix was not symmetric (dist[j,i] = dist[i,j])

def get_possibilities(i):
    return [j for j in [i+1, i-1, i - n, i + n, i - n - 1, i - n + 1, i + n - 1, i + n + 1] if j >= 0 and j < N 
            and (i % n != n-1 or j % n != 0) and (i % n != 0 or j % n != n-1)]

def get_income(source, dest):
    source_x = source // n
    source_y = source % n
    dest_x = dest // n
    dest_y = dest % n
    return math.sqrt((source_x - dest_x)**2 + (source_y - dest_y)**2)

######### PROBLEM PARAMETERS 

number_of_cars = 100
n = 4 # number of cells sqrt
N = n*n
WC = 1 # walk cost

#epsilon = 0.1 -- no more epsilon, it was the probability of finding/not finding the customer
delta = 0.5

r = np.ones((number_of_cars, N))*(1/N)
init_car_locations = [np.random.randint(N) for l in range(number_of_cars)]
p = np.zeros((number_of_cars, N))
for i in range(number_of_cars):
    p[i,init_car_locations[i]] = 1

######### CREATING DEMAND 

random_average = np.random.random((N,N))*0.1
np.fill_diagonal(random_average, 0)
# decrease income and outcome of a cell
random_average[:,2] *= 0.0001
random_average[4,:] *= 20
random_average[:,9] *= 10
random_average[:,14] *= 30

######### CALCULATING CONSTANT MATRICES
expected_demand = np.sum(random_average,axis=1) # axis=1 => summation of TO values
Pr = (random_average.T / expected_demand).T

dist = np.zeros((N,N))
for i in range(N):
    for j in range(i,N):
        dist[i,j] = get_income(i,j)
        dist[j,i] = dist[i,j]

variables = []
for i in range(N):
    possibilities = get_possibilities(i)
    for possibility in possibilities:
        variables.append((i, possibility))

A = np.zeros((N,N))# transition matrix for random walks, a policy, to make sparse #TODO
for i in range(N):
    possibilities = get_possibilities(i)
    for j in possibilities:
        A[i][j] = 0.5
A = (A.T / np.sum(A,axis=1)).T
```

```python
########## SIMULATIONS

def run_simulations(policy, random_average, time_periods):

    # now we define probabilities for each transition
    # we assume poisson distribution
    # ew learn poisson distribution by fitting lambda to the average number of events in the cell

    income_per_car = np.zeros(number_of_cars)
    trips_per_car = np.zeros(number_of_cars)

    ###########
    cell_to_cars = {}
    for i in range(N):
        cell_to_cars[i] = []
    for l in range(number_of_cars):
        cell_to_cars[init_car_locations[l]].append(l)
    cars_to_cell = {}
    for i in range(number_of_cars):
        cars_to_cell[i] = init_car_locations[l]

    ########
    for t in range(time_periods):
        trips_per_cell = np.random.poisson(random_average)
        next_cars_to_cell = {}
        # distribute people
        for cell in range(N):
            m = min(sum(trips_per_cell[cell]), len(cell_to_cars[cell]))
            if m == 0:
                continue
            assined_cars = np.random.choice(cell_to_cars[cell], m, replace=False)

            car_id = 0
            for i in np.random.choice(N, N, replace=False):
                while trips_per_cell[cell][i] > 0 and car_id < len(assined_cars):
                    next_cars_to_cell[assined_cars[car_id]] = i
                    income_per_car[assined_cars[car_id]] += get_income(cell, i)
                    trips_per_car[assined_cars[car_id]] += 1
                    car_id += 1
                    trips_per_cell[cell][i] -= 1
        # assign non-assigned
        for i in range(N):
            cell_to_cars[i] = []

        for i in range(number_of_cars):
            if i not in next_cars_to_cell:

                # here we apply policy
                next_cars_to_cell[i] = policy(cars_to_cell[i])
                income_per_car[i] -= WC
            cell_to_cars[next_cars_to_cell[i]].append(i)

        cars_to_cell = next_cars_to_cell
    return income_per_car, trips_per_car, cars_to_cell

############ RUNNING MC SIMULATIONS
def get_mc_results():
    visits = np.zeros((number_of_cars, N))
    objective = np.zeros(number_of_cars)
    mc_it = 50
    walk_periods = 50
    for mc_sim in range(mc_it):
        results = run_simulations(A_policy, random_average, walk_periods)
        for i in range(number_of_cars):
            visits[i,results[2][i]] += 1
        objective += results[0]
    objective /= mc_it
    return (np.min(objective), np.mean(objective), np.max(objective))
```

```python
######### POLICIES

def where_to_go(i):
    # select one of 9 neighbors randomly
    possibilities = get_possibilities(i)
    return np.random.choice(possibilities)

def where_to_go_smart(i):
    possibilities = get_possibilities(i)
    i = np.argmax(random_average_source[possibilities])
    return possibilities[i]

def A_policy(i):
    return np.random.choice(range(A.shape[1]),p=A[i,:])
```

```python
########## RUNNING RANDOM WALK

ser = []
for t in range(100):
    expected_number_of_cars_per_cell = np.sum(p, axis=0)
    probability_of_busy = np.minimum(1, np.divide(expected_demand, expected_number_of_cars_per_cell))

    B = (Pr.T * probability_of_busy).T + (A.T * (1-probability_of_busy)).T
    phi = np.dot(p, B)

    r_next = (1-delta)*r + delta*phi
    p_next = r_next
    p = p_next
    r = r_next
    ser.append(p[0][:3])
```

```python
conv_p = []
conv_a = []
all_results = []

A = np.zeros((N,N))# transition matrix for random walks, a policy, to make sparse #TODO
for i in range(N):
    possibilities = get_possibilities(i)
    for j in possibilities:
        A[i][j] = 0.5
A = (A.T / np.sum(A,axis=1)).T

for t in tqdm(range(75)):
    expected_number_of_cars_per_cell = np.sum(p, axis=0)
    probability_of_busy = np.minimum(1, np.divide(expected_demand, expected_number_of_cars_per_cell))

    B = (Pr.T * probability_of_busy).T + (A.T * (1-probability_of_busy)).T
    phi = np.dot(p, B)
    r_next = (1-delta)*r + delta*phi
    p_next = r_next
    p = p_next
    r = r_next
    conv_p.append(p[0][0])
    
    #optimizing
    m = Model('taxi')
    m.setParam('OutputFlag', 0)

    z = m.addVar(lb=-GRB.INFINITY, ub=GRB.INFINITY, vtype=GRB.CONTINUOUS, name="z")
    policy = m.addVars(variables, lb=0.0, ub=1.0, name="policy", vtype=GRB.CONTINUOUS)
    m.addConstrs((policy.sum(i,'*') == 1 for i in range(N)), "cap")
    m.setObjective(z, GRB.MAXIMIZE)

    expected_number_of_cars_per_cell = np.sum(p, axis=0)
    probability_of_busy = np.minimum(1, np.divide(expected_demand, expected_number_of_cars_per_cell))
    H = np.sum(np.multiply(Pr, dist), axis=1) * probability_of_busy - (WC * (1-probability_of_busy))

    coef_slag = np.dot( np.dot(p, (Pr.T * probability_of_busy).T), H)
    coef_dict = []
    for q in range(number_of_cars):
        coef_dict.append({})
        for var in variables:
            k = var[0]
            l = var[1]
            coef_dict[-1][var] = float(p[q, k] * (1-probability_of_busy[k]) * H[l])
    costr = m.addConstrs((policy.prod(coef_dict[q]) + coef_slag[q] - z >= 0 for q in range(number_of_cars)), "cap")
    m.optimize()
    
    solution = m.getAttr('x', policy)
    beta = 0.1
    for var in solution:
        A[var[0], var[1]] = A[var[0], var[1]]+solution[var]*beta
    A = (A.T / np.sum(A,axis=1)).T
    conv_a.append(A[0,1])
    
    if t % 10 == 0:
        all_results.append((t, get_mc_results()))
```
